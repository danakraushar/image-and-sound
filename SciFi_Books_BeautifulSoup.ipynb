# Web Scrape of Books by Female Sci-Fi Authors, from GoodReads.com 


Scraped data on the top-rated sci-fi books written by female authors, via GoodReads.com. Compiled .csv table including rank, book title, book URL, full rating text, and full score text. Scraping done using Beautiful Soup.

```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.ScienceFictionBooksbyFemale_Authors')
doc4 = BeautifulSoup(response.text, 'html.parser')
```

```python
data4 = []
elements = doc4.find_all('tr', {'itemtype': 'http://schema.org/Book'})
for element in elements:
    print(element)
    datapoint = {}
    datapoint['rank'] = element.find(class_='number').text.strip()
    datapoint['book_title'] = element.find(class_="bookTitle").text.strip()
    datapoint['book_URL'] = element.find('a')['href']
    datapoint['full_score_text'] = element.find('a', onclick=True).text.strip()
    datapoint['full_rating_text'] = element.find(class_='minirating').text.strip()
    data4.append(datapoint)
pd.DataFrame(data4)
df4 = pd.DataFrame(data4)
df4.to_csv('sci-fi.csv', index=False)
```
