```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd

%matplotlib inline
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf')
```

```python
data = []
articles = driver.find_elements_by_class_name('titlebet')
for article in articles:
    datapoint = {}
    datapoint['headline'] = article.text
    print(article.text)
    datapoint['onclick'] = article.get_attribute('onclick')
    print(article.get_attribute('onclick'))
    data.append(datapoint)
```

```python
df = pd.DataFrame(data)
df.to_csv('nk-news-selenium2.csv', index=False)
```
