# Web Scrape of H&M's website

I scraped H&M's website's (hm.com) home section to identify items on sale and compiled the product names, old and new prices, and url's into a .csv file. Scraping performed using Selenium.

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd

%matplotlib inline
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('http://www.hm.com/us/products/sale/home')
```

```python
data5 = []
products = driver.find_elements_by_class_name('product-list-item')
for product in products:
#     print(product.find_element_by_class_name('product-title').text)
#     print(product.find_element_by_class_name('old-price').text)
#     print(product.find_element_by_class_name('price').text)
#     print(product.find_element_by_tag_name('a').get_attribute('href'))
    datapoint = {}
    datapoint['name'] = product.find_element_by_class_name('product-title').text
    datapoint['old_price'] = product.find_element_by_class_name('old-price').text
    datapoint['new_price'] = product.find_element_by_class_name('price').text
    datapoint['url'] = product.find_element_by_tag_name('a').get_attribute('href')
    data5.append(datapoint)
df5 = pd.DataFrame(data5)
df5.to_csv('hm-sales.csv', index=False)
```
